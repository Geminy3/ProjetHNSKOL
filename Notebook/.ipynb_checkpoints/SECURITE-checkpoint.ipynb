## Désambiguïsation et analyse du terme "sécurité" 

Les analyses qui suivent visent à remédier à l'ambiguïté du terme "sécurité" constatée dans le corpus et rendant complexe l'interprétation des résultats précédemment obtenus. Un traitement manuel du corpus a été réalisé, visant à désambiguïser le terme "sécurité" en remplaçant chacune de ses occurrences par un unique token composé du terme "securite" et d'un qualificatif précisant le type de sécurité auquel il est fait référence. 

Les principales notions recouvertes par le terme "sécurité" dans ce corpus sont les suivantes : 

- La collocation **sécurité nationale** renvoie à la protection des intérêts nationaux essentiels. Elle apparaît dans le corpus sous la forme du token *securitenationale*
- La **sécurité intérieure**, quant à elle, est une des fonction régaliennes de l'Etat et s'applique aussi bien aux biens qu'aux personnes. Elle comprend la **sécurité publique**, terme qui renvoie à tous les dipositifs visant à protéger les personnes et les biens contre les crimes, la délinquance et les troubles à l'ordre publique, et la **sécurité civile**, désignant les dispositifs de prévention à l'encontre des accidents et catastrophes naturelles et technologiques. On observe cependant, du fait d'un glissement sémantique, une superposition des qualificatifs "intérieure" et "publique" pour désigner la sécurité des personnes et des biens faces aux menaces criminelles, dont le terrorisme. Il a donc été décidé d'utiliser un unique token pour désigner ces deux notions régulièrement confondues : *securiteinterieure*. La sécurité civile, quant à elle, demeure distincte et apparaît, à ce titre, sous la forme d'un token distinct : *securitecivile*.
- La **sécurité extérieure** renvoie en revanche à la protection d'un Etat face aux menaces extérieures. Le terme **défense** s'y substitue régulièrement. Il a été décidé de ne pas regrouper les deux expressions sous un même token afin de ne pas dénaturer outre mesure le corpus. Nous tiendrons cependant compte de leur quasi-synonymie lors de l'analyse des occurrences des tokens *securiteexterieure* et *défense*. A noter cependant : la loi d'orientation et de programmation pour la performance de la sécurité intérieure datant de 2011 visait à abolir la dichotomie entre les notions de défense (sécurité extérieure) et de sécurité intérieure. 
- L'expression **sécurité internationale**, apparaissant dans le texte sous la forme du token *securiteinternationale*, a été choisie pour désigner tous les dispositifs visant à renforcer la protection d'un ou plusieurs Etats autres que la France : elle désigne notamment, dans notre corpus, la sécurité du continent africain - thématique majeure de l'action militaire française à l'étranger ces dernières années. 
- La collocation **sécurité sanitaire**, qui désigne les dispositifs visant à protéger une population des risques pour leur santé, n'apparaît aucunement dans le corpus, contrairement à ce que l'on aurait pu attendre de projets de loi publiés après 2020.
- La **sécurité alimentaire** désigne les dispositifs visant à garantir l'accès d'une population à une alimentation de qualité et en quantité suffisante. Elle apparaît dans le corpus sous la forme du token *securitealimentaire*.
- La **sécurité juridique** vise la protection des droits et libertés publiques et apparaît dans le corpus sous la forme du token *securitejuridique*. 
- La **sécurité financière** désigne tous les dispositifs visant à protégrer les flux financiers, et apparaît sous la forme du token *securitefinanciere*
- La **sécurité du travail** peut faire référence à la prévention des risques professionnels (accidents du travail, risques chimiques, électriques, psychosociaux ou liés à l'activité physique) ou bien à la protection de salariés contre le licenciement abusif. Nous regrouperont ces différents sèmes sous un seul token, *securitetravail. 
- La **sécurité des transports** est une notion qui englobe la sécurité routière, ferrovière, aérienne et maritime, et apparaît donc dans le corpus sous la forme d'un unique token : *securitetransports*.
- La **sécurité informatique** ou **cybersécurité**, désigne l'ensemble des dispositifs visant à protéger les systèmes d'information, et apparaît dans le corpus sous la forme du token *securiteinformatique*
- La **sécurité des produits** désigne l'ensemble des dispositifs visant à contrôler le respect des normes nationales ou internationales des produits circulant sur le territoire national. Elle comprend à ce titre la sécurité des médicaments et des soins et apparaît sous la forme du token *securiteproduits*.
- La **sécurité énergétique** désigne la disponibilité des ressources naturelles pour la consommation d'énergie, et apparaît sous la forme du token *securiteenergetique*. Elle se distingue de la **sécurité nucléaire** dans la mesure où cette dernière faire surtout référence à l'ensemble des mesures relatives à la protection des matières nucléaires, de leurs installations et de leur transport contre des actes malveillants, permettant de garantir l'absence de conséquence inacceptable pour les populations et l'environnement. La sécurité nucléaire peut, à ce titre, être rapprochée de la sécurité civile. Il a cependant été décidé de l'en distinguer par le token *securitenucleaire*, en raison de sa spécificité. 
- La collocation **sécurité sociale**, enfin, désigne généralement l'ensemble des institutions ayant pour vocation la protection des individus contre l'ensemble des risques sociaux, et apparaît dans le corpus sous la forme du token *securitesociale*.

Par ailleurs, le terme "sécurité" apparaît également en tant que concept général dans le corpus, sans qu'il ne fasse référence à un type de sécurité en particulier, auquel cas il a été remplacé par le token générique *securitegenerale*.

Ce travail de désambiguïsation a consisté, dans un premier temps, à agréger les collocations les plus courantes afin de former les tokens suscités (par exemple : securitesociale, securitenationale, etc.) et, dans un deuxème temps, à identifier le type de sécurité auquel font référence les occurrences isolées du terme "sécurité" en s'appuyant sur leur contexte direct, dans le but de les remplacer par un token permettant d'en préciser le sens.

```python
def top_feats(row, features, top_n=25):
    ''' Get top n tfidf values in row and return them with their corresponding feature names.'''
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids if row[i]>0]
    df = pd.DataFrame(top_feats)
    if len(top_feats) > 0:
        df.columns = ['feature', 'score']
    return df
```
